```python
# First, import the relevant modules
import requests
from collections import OrderedDict
import numpy as np
```

Note: API's can change a bit with each version, for this exercise it is reccomended to use the "V3" quandl api at `https://www.quandl.com/api/v3/`

```python
# Now, call the Quandl API and pull out a small sample of the data (only one day) to get a glimpse
# into the JSON structure that will be returned
url = 'https://www.quandl.com/api/v3/datasets/XFRA/AGB2_UADJ?start_date=2017-01-01&end_date=2017-12-31&api_key='
r = requests.get(url)
json_data = r.json()
json_data
```

```python
# Inspect the JSON structure of the object you created, and take note of how nested it is,
# as well as the overall structure
json_data['dataset'].keys()
```

These are your tasks for this mini project:

1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).
2. Convert the returned JSON object into a Python dictionary.
3. Calculate what the highest and lowest opening prices were for the stock in this period.
4. What was the largest change in any one day (based on High and Low price)?
5. What was the largest change between any two days (based on Closing Price)?
6. What was the average daily trading volume during this year?
7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
AFX = json_data['dataset']
dates_trades = AFX['data'][:-1]
dates_trades
```

```python
column_names = json_data['dataset']['column_names']
column_names
```

Used list comprehension to extract trade dates for 2017 from json_data

```python
dates = [item[0] for item in dates_trades]
dates
```

Used list comprehension to extract open prices, high prices, low prices, closing prices, and stock volume for each date during year 2017. This was stored in variable 2017. 

```python
sp_2017 = [item[1:6] for item in dates_trades]
sp_2017 = np.array(sp_2017)
sp_2017
```

Created a dictionary to match dates with stock price information. Date was used as the key while open prices, high prices, low prices, closing prices, and stock volume were used as values in the dictionary. The dictionary was stored in variable date_to_sp as shown below.

```python
date_to_sp = dict(zip(dates, sp_2017))
date_to_sp
```

```python
# creating variables for stock price and stock volume for later use to answer questions 3, 4, 5, 6, and 7
closing_price = sp_2017[:,3]
volume = sp_2017[:,4]
open_price = sp_2017[:,0]
```

```python
#question 3
highest_open_2017 = np.max(open_price)
highest_open_2017
```

```python
#question 3 continued..
lowest_open_2017 = np.min(open_price)
lowest_open_2017
```

```python
#question 4 
diff_high_low = np.max(sp_2017[:,1] - sp_2017[:,2])
diff_high_low
```

```python
#question 5
diff_close_two_days = [abs(x - y) for x, y in zip(closing_price[1:], closing_price[:-1])]
diff_close_two_days
max_diff_close_two_days = max(diff_close_two_days)
max_diff_close_two_days
```

```python
#question 6
avg_volume = np.mean(volume)
avg_volume
```

```python
#question 7
median_volume = np.median(volume)
median_volume
```
